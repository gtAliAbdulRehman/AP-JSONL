```python
import pandas as pd
import numpy as np
import json

dt = pd.read_excel('NonMaths.xlsx', sheet_name = 'Non-Math Core Skills')
dt.head()
```

```python
dtl1 = dt.drop_duplicates(subset=["Standard Id (L1)", "Standard Description (L1)"])
```

```python
dtl1
```

```python
##FOR L1
dataL1 = []
for i in dtl1.itertuples(index = False):
    prompt = f"What is {i[1]} with Domain: {i[3]}, Cluster: {i[4]}  and Standard: {i[5]} ?"
    completion = i[6]
    dataL1.append({"prompt": prompt, "completion":completion})

with open("NMA_L1.jsonl", "w") as L1:
    json.dump(dataL1, L1)
```

```python
with open('NMA_L1.jsonl', "r") as f:
    dt_l1 = json.load(f)

with open('NMA_L1.jsonl', "w") as f:
    for i in range(len(dt_l1)):
        json.dump(dt_l1[i], f)
        if i != len(dt_l1) - 1:
            f.write('\n')
```
