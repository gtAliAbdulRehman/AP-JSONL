```python
import pandas as pd
import numpy as np
import json

dt = pd.read_excel('CBS.xlsx', sheet_name = 'College Board Standards', header = 5)
```

```python
dt.tail()
```

```python
dt.drop('Big Idea', axis =1, inplace = True)
dt.head()
```

```python
dt['Essential Knowledge'] = np.where(dt['Essential Knowledge'].isnull() | (dt['Essential Knowledge'] == 'None'), dt['Learning Objective'], dt['Essential Knowledge'])
```

```python
dt.head()
```

```python
dt.dropna(inplace = True)
```

```python
dt.head()
```

```python
dt.isnull().sum()
```

```python
dt.shape
```

```python
dt.tail()
```

```python
data = []
for i in dt.itertuples(index = False):
    prompt = f"What is Mod {i[2]} in AP Computer Science with Topic: {i[3]}, and Summary: {i[4]} ?"
    completion = i[6]
    data.append({"prompt": prompt, "completion":completion})

with open("CBR_All.jsonl", "w") as result:
    json.dump(data, result)
```

```python
with open('CBR_All.jsonl', "r") as f:
    dt = json.load(f)

with open('CBR_All.jsonl', "w") as f:
    for i in range(len(dt)):
        json.dump(dt[i], f)
        if i != len(dt) - 1:
            f.write('\n')
```
